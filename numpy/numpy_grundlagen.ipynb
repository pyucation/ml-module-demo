# Numpy Grundlagen

## Was ist Numpy?
Numpy (NumPy) bedeutet "Numerical Python" und ist eine Bibliothek, die vor allem für mathematische Berechnungen ungeheime Geschwindigkeitsvorteile bietet. Numpy's **arrays** sind bis zu 50-mal schneller als Python's Listen*. Vor allem im Data Science Bereich, wo wir mit großen Datenmengen arbeiten, ist das ein entscheidendes Kriterium.\
Numpy's eingebaute Funktionen für diese `ndarray`s sind einfach anzuwenden und sind (oft) leicht zu verstehen. Im Folgenden werden wir uns die Grundlagen dieser `ndarray`s anschauen.

<sub>* Das liegt an der Art und Weise, wie die Daten im Computer gespeichert werden. Außerdem sind viele rechenaufwendige Berechnungen in C oder C++ geschrieben, was einen deutlichen Geschwindigkeitsvorteil gegenüber Python bietet.</sub>

Beginnen wir wie immer mit dem Import:

```python
import numpy as np
```

Wir können eine Python-Liste ganz einfach in ein `ndarray` umwandeln:

```python
py_list = [3, 4, 1, 9]
np_array = np.array(py_list)

print(np_array)
print(type(np_array))
```

Ein wichtiges Konzept sind die Achsen (axes) eines `ndarray`s. Sie entsprechen der Anzahl der (mathematischen) Dimensionen.
- `ndim`: Anzahl der Achsen (Dimensionen)
- `shape`: Dimensionen des Arrays, also die Größe in jeder Dimension
- `size`: Anzahl der Elemente

```python
# ein Skalar (0-dim)
scalar = np.array(7)
print(f"ndim: {scalar.ndim}")
print(f"shape: {scalar.shape}")
print(f"size: {scalar.size}")
```

```python
# ein 1D-Array
arr_1D = np.array([3, 9, 5, 1])
print(f"ndim: {arr_1D.ndim}")
print(f"shape: {arr_1D.shape}")
print(f"size: {arr_1D.size}")
```

```python
# ein 2D-Array
arr_2D = np.array([[1, 2, 3], [4, 5, 6]])
print(f"ndim: {arr_2D.ndim}")
print(f"shape: {arr_2D.shape}")
print(f"size: {arr_2D.size}")
```

```python
# ein 3D-Array
arr_3D = np.array([[[1, 2, 3], [4, 5, 6]], [[99, 98, 97], [96, 95, 94]]])
print(f"ndim: {arr_3D.ndim}")
print(f"shape: {arr_3D.shape}")
print(f"size: {arr_3D.size}")
```

Du kannst die Dimension beim Erstellen eines Arrays angeben:

```python
arr = np.array([0, 1, 1, 2, 3, 5, 8], ndmin=4)

print(arr)
```

## Verschiedene Arten `ndarray`s zu erzeugen

Bisher haben wir gelernt, wie wir eine Python-Liste in ein numpy array umwandeln. Wir können jedoch auch numpy direkt ein array erzeugen lassen. Da wir in der Praxis die Elemente eines Arrays oftmals nicht kennen (bspw. die Gewichte eines Neuronalen Netzes), rüstet Numpy uns mit Platzhaltern aus:

```python
# erzeugt ein 3x3 array mit Nullen
zeros = np.zeros((3, 3))
print(zeros)
```

```python
# erzeugt ein 2x4 array mit Einsen
ones = np.ones((2, 4))
print(ones)
```

```python
# erzeugt ein 2x2 Array mit Zahlen, ohne sie zu initialisieren
empty = np.empty((2, 2))
print(empty)
```

```python
# erzeugt ein Array der angegeben shape (hier 3x3) mit dem gegebenem "fill_value"
filled = np.full((3, 3), fill_value=5) # fill_value kann auch ein Array/List sein
print(filled)
```

Ähnlich wie die `range`-Funktion in Python, bietet Numpy Funktionen zum Erzeugen von Zahlen:
- `np.arange(start, stop, schrittweite)`, *stop* ist nicht eingeschlossen

Für Fließkommazahlen-Schritte (z.B. 0.1) ist es besser die folgende Funktion zu verwenden:
- `np.linspace(start, stop, anzahl)`, *stop* ist eingeschlossen

```python
x = np.arange(0, 10, 2)
print(f"arange: {x}")
```

```python
x = np.linspace(0, 10, 5)
print(f"linspace: {x}")
```

## Indexing Numpy Arrays

Wie wir es aus Python gewohnt sind, können wir mit eckigen Klammern [ ] auf die Elemente eines Arrays zugreifen.\
Numpy bietet bei mehrdimensionalen Arrays, auf die vielen eckigen Klammern zu verzichten und die Dimensionen stattetdessen mit Kommas zu trennen.

```python
x = np.array([[1, 2], [3, 4]])
print(x[1][1]) # so kennen wir es aus Python
print(x[1, 1]) # ebenfalls möglich in Numpy
```

Ein negativer Index zählt, wie in Python, von hinten:

```python
print(x[0, -1]) # das letzte Element (-1) der ersten (0) Dimension (Zeile)
```

## Slicing

Slicing in Numpy funktioniert grundlegend wie in Python.

```python
x = np.array([[3, 7, 5], [2, 0, 1]])

print(f"Zeile 2, ab zweitem Element: {x[1, 1:]}") # Dimension 1, ab dem zweiten Element bis zum Ende
print(f"komplette Spalte 1: {x[:,0]}")
print(f"komplette Zeile 2: {x[1,]}")
```

## Reshape

*reshape* bedeutet, dass wir das `shape` Atttribut eines Arrays verändern. Erinnerung: `shape` ist die Anzahl der Elemente eines Arrays in jeder Dimension.

```python
# ein 1D Array mit 10 Elementen
x = np.arange(0, 10)
print(x.shape)
```

```python
# reshape zu einem 2D Array mit 2 Zeilen und 5 Spalten
x_new = x.reshape(2, 5)
print(x_new.shape)
print(x_new)
```

Beachte: Die Anzahl der Elemente nach dem *Reshape* muss gleich der Anzahl der Elemente vor dem *Reshape* sein.

```python
# 1D Array mit 8 Einträgen
x = np.arange(0, 8)
print(f"original x: {x}\n")

# 2D Array mit 2x4 Dimensionen, 2*4 = 8
x_new_2D = x.reshape(2, 4)
print(f"reshaped 2D: {x_new_2D}\n")

# 3D Array mit 2x2x2 Dimensionen, 2*2*2 = 8
x_new_3D = x.reshape(2, 2, 2)
print(f"reshaped 3D:\n{x_new_3D}")

# 3D Array mit 3x2x2 Dimensionen, 3*2*2 = 12 != 8
_ = x.reshape(3, 2, 2)
```

Beachte: `reshape` gibt keine Kopie zurück. Änderungen werden genauso auf das originale Array angewendet.

```python
x_new_2D[1, 1] = 99 # wir ändern x_new_2D
print(x) # und sehen die Änderungen auch in x
```

Wir dürfen auch eine *unbekannte* Anzahl an Dimensionen haben. Dazu schreiben wir `-1`. Numpy berechnet die Dimension selbst.

```python
x_unbekannt = x.reshape(2, -1)
print(x_unbekannt)
```

Das ist vor allem nützlich, wenn wir aus einem Zeilenvektor (eine einfache Liste/1D Array) einen Spaltenvektor machen müssen (und andersherum).

```python
zeilen_vec = np.array([4, 9, 1, 2])
spalten_vec = zeilen_vec.reshape(-1, 1)
print(spalten_vec)
```

## Iterieren über `ndarray`

Wir können `for`-loops, die wir aus Python kennen, nutzen. Allerdings wird dies schnell kompliziert (da wir für n Dimensionen n for loops benötigen), wenn wir mehr als zwei oder drei Dimensionen haben. Stattdessen stellt Numpy die Funktion `nditer()` zur Verfügung.

```python
arr = np.array([[6, 5], [4, 3], [2, 1]])

# herkömmliches Python
for row in arr:
  for element in row:
    print(element)
```

```python
# mit nditer
for element in np.nditer(arr):
  print(element)
```

Falls wir ebenso den Index des Elements benötigen nutzen wir `ndenumerate` (vergleichbar mit Python's `enumerate`):

```python
for index, element in np.ndenumerate(arr):
  print(index, element)
```

## Mehrere Arrays zusammenfügen

Dafür gibt es mehrere Funktionen, jede für einen unterschiedlichen Anwendungsfall:
- `concatenate`: verknüpft zwei Arrays entlang einer **bestehenden** gegebenen Achse (0, wenn nichts gegeben wurde)
- `stack`: verknüpft zwei Arrays entlang einer **neuen** gegebenen Achse (0, wenn nichts gegeben wurde)
- `hstack`: entlang einer Zeile verknüpfen
- `vstack`: entlang einer Spalte verknüpfen

```python
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([5, 6, 7, 8])
```

```python
# concatenate
new_arr = np.concatenate((arr1, arr2))
print(new_arr)
```

```python
# concatenate entlang einer Zeile
new_arr = np.concatenate((arr1.reshape(2, 2), arr2.reshape(2, 2)), axis=1) # Übung: was passiert hier?
print(new_arr)
```

```python
# stack
new_arr = np.stack((arr1, arr2))
print(new_arr)
```

```python
new_arr = np.stack((arr1, arr2), axis=1)
print(new_arr)
```

```python
# hstack
new_arr = np.hstack((arr1, arr2))
print(new_arr)
```

```python
# vstack
new_arr = np.vstack((arr1, arr2))
print(new_arr)
```

## Suchen in Arrays

Zum Suchen können wir zwei Funktionen benutzen:
- `where()`: liefert die Indices zurück, wo die Suche "erfolgreich" war
- `searchsorted()`: Array muss sortiert sein, liefert den ersten Index wo die Suche "erfolgreich" war, binäre Suche

```python
arr = np.array([5, 8, 1, 9, 4, 2, 1])
```

```python
# where
indices = np.where(arr == 1)
print(indices)
```

```python
# searchsorted
arr.sort()
wo = np.searchsorted(arr, 9)
print(wo)
```

## Zufallsverteilungen und Zufallswerte in Numpy

Oft müssen wir mit Zufallswerten arbeiten. In Computern können wir jedoch nur bedingt wirklich zufällige Werte erzeugen. Es gibt Wege dies zu tun, die vor allem für sicherheitsrelevante Anwendungen genutzt werden. Wir werden jedoch mit sogenannten "Pseudo-Zufallszahlen" arbeiten.

```python
# eine zufällige Ganzahl (Integer) zwischen 0 und 10
x = np.random.randint(10)
print(x)
```

```python
# eine zufällige Fließkommazahl (Float) zwischen 0 und 1
x = np.random.rand()
print(x)
```

Wir können auch ein Array mit diesen beiden Funktionen erzeugen. Dafür geben wir eine `size` an. Beachte: Numpy ist hier etwas inconsistent, `size` entspricht in dem Fall einer `shape`.

```python
rand_arr = np.random.randint(10, size=(2, 3))
print(rand_arr)
```

```python
rand_float_arr = np.random.rand(2, 2)
print(rand_float_arr)
```

Mit der `choice()` Funktion kann man aus einer Liste von Werten zufällig einen Wert auswählen.

```python
my_choice = np.random.choice([5, 3, 0, 9])
print(my_choice)
```

Das ist auch als Array möglich:

```python
my_arr_choice = np.random.choice([5, 3, 0, 9], size=(2, 3))
print(my_arr_choice)
```

Ohne eine bestimmte Wahrscheinlichkeitsverteilung anzugeben, wählt Numpy die Werte gleichverteilt (jeder Wert hat die gleiche Wahrscheinlichkeit gezogen zu werden) aus. Wir können über den Parameter `p` (engl.: *probability*) eine Verteilung angeben:

```python
my_choices = np.random.choice([5, 3, 0, 9], p=[0.2, 0.7, 0.05, 0.05], size=(20))
print(my_choices)
```

Werte aus einer Normalverteilung mit Mittelwert $\mu$ und Standardabweichung $\sigma$ erhalten wir über die Funktion `random.normal()`. Dabei entspricht `loc` dem Mittelwert und `scale` der Standardabweichung.

```python
normal_values = np.random.normal(loc=2, scale=2.5, size=(1000))

# zum Plotten (wird hier nicht weiter besprochen)
import seaborn as sns


sns.distplot(normal_values)
```

Ebenso gibt es auch eine Gleichverteilung (engl.: *uniform distribution*) und weitere Verteilungen.

```python
uniform_values = np.random.uniform(0, 1, size=(3, 3))
print(uniform_values)
```
